```python
import pandas as pd
import os
from pathlib import Path
```

```python
def rr_clean(df):
    df = df.rename(columns={' State': 'State', ' Sample': 'Sample', ' Replacement Ratio 1': 'Replacement Ratio 1',
       ' Replacement Ratio 2': 'Replacement Ratio 2', ' Average WBA': 'Average WBA', ' Average Weekly Wage': 'Average Weekly Wage'})
    df = df.loc[:,['Year', 'State', 'Replacement Ratio 1', 'Replacement Ratio 2', 'Average WBA', 'Average Weekly Wage']]
    return df
```

```python
def add_to_avgwba_aww_rr():
    
    # Path to folder
    cwd = os.getcwd()
    folder_path = Path(cwd + "/raw")

    # List to store individual DataFrames
    dfs = []

    # Loop through all Excel files
    for file in folder_path.glob("*.csv"):
        df = pd.read_csv(file)
        df = rr_clean(df)
        dfs.append(df)

    # Append all rows together
    current = pd.read_csv("avgwba_aww_rr.csv")
    dfs.append(current)
    final_df = pd.concat(dfs, ignore_index=True)
    final_df = final_df.dropna()
    final_df = final_df.sort_values(by=["Year", "State"], ascending=True)
    
    final_df.to_csv("avgwba_aww_rr.csv", index=False)
    
    return
```

```python
add_to_avgwba_aww_rr()
```

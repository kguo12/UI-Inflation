```python
import pandas as pd
from pathlib import Path
import os
```

## Add Labor Force Data

```python
## INSERT year, laborforce dataframe
laborforce = pd.read_csv("laborforce2003-2023.csv")
year = 2024
add_year_to_laborforce(year, laborforce)
```

```python
def add_year_to_laborforce(year, laborforce):
    
    # Path to folder
    cwd = os.getcwd()
    folder_path = Path(cwd + f"/{year}raw")

    # List to store individual DataFrames
    dfs = []

    # Loop through all Excel files
    for file in folder_path.glob("*.csv"):
        df = pd.read_csv(file)
        dfs.append(df)

    # Append all rows together
    final_df = pd.concat(dfs, ignore_index=True)
    final_df = final_df.dropna()

    laborforce = pd.concat([laborforce, final_df], ignore_index=True)
    laborforce.to_csv(f"laborforce{year}.csv")
    
    return
```

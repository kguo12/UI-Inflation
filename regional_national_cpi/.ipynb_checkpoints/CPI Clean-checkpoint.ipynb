### 1. National and Regional XLSX data -> `cpi.csv`

```python
# Path to folder
cwd = os.getcwd()
folder_path = Path(cwd + "/raw")

dfs = []

# Loop through all Excel files
for file in folder_path.glob("*.xlsx"):
    region = re.search(r'^([A-Za-z]+)', Path(file).name).group(1)
    df = pd.read_excel(file)
    df = df.loc[11:]
    df = df[['Consumer Price Index for All Urban Consumers (CPI-U)', 'Unnamed: 13']]
    df = df.rename(columns={'Consumer Price Index for All Urban Consumers (CPI-U)': "Year", 'Unnamed: 13': f"{region}"})
    df.to_csv(f'{file}.csv', index=False)
    dfs.append(df)

merged = reduce(lambda left, right: pd.merge(left, right, on="Year", how="outer"), dfs)
merged["National"] = merged.pop("National")
merged.to_csv("cpi.csv", index=False)
```

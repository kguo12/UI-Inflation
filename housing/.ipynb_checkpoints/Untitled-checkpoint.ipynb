```python
import pandas as pd
```

```python
housing2010 = pd.read_csv("housing2010.csv")
housing2023 = pd.read_csv("housing2023.csv")
```

```python
housing2010 = housing2010.dropna()
housing2010['Year'] = 2010
housing2023 = housing2023.dropna()
housing2023['Year'] = 2023
```

```python
housingcosts = pd.concat([housing2010, housing2023], ignore_index=True)
housingcosts
```

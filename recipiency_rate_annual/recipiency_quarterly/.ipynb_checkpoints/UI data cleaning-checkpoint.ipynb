```python
import pandas as pd
```

```python
def rr_clean(df):
    df = df.iloc[:-1]
    df = df.rename(columns={' State': 'State', ' Sample': 'Sample', ' Replacement Ratio 1': 'Replacement Ratio 1',
       ' Replacement Ratio 2': 'Replacement Ratio 2', ' Average WBA': 'Average WBA', ' Average Weekly Wage': 'Average Weekly Wage'})
    df = df.loc[:,['Year', 'State', 'Replacement Ratio 1', 'Replacement Ratio 2', 'Average WBA', 'Average Weekly Wage']]
    return df
```

```python
import glob
import os

path = r"C:\Users\Prophecy Desktop\OneDrive\Documents\GitHub\UI-Inflation\replacement_rates"
all_files = glob.glob(os.path.join(path , "*.csv"))
all_files

li = []

for filename in all_files:
    df = pd.read_csv(filename, delimiter=", ", index_col=None, header=0)
    df = rr_clean(df)
    li.append(df)

rr_wba_wage = pd.concat(li, axis=0, ignore_index=True)
```

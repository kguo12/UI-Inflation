Real Price = Nominal Price * (CPI in Base Period / CPI in Current Period)

```python
base2000_cpis = {
            "West": 174.8,
            "South": 167.2,
            "Northeast": 179.4,
            "Midwest": 168.3
            }

base2023_cpis = {
            "West": 323.834,
            "South": 296.422,
            "Northeast": 313.786,
            "Midwest": 282.76
            }


def get_regional_real2000(data, column):
    base = base2000_cpis[data['Region']]
    current = data['Annual Regional CPI-U']
    real_amount = round(data[column] * (base/current), 2)
    return real_amount

def get_regional_real2023(data, column):
    base = base2023_cpis[data['Region']]
    current = data['Annual Regional CPI-U']
    real_amount = round(data[column] * (base/current), 2)
    return real_amount
```

```python
def calc_state_cpis(row, data):
    year_index = row['Year'] - 2000
    state = row['State']
    cpi = 100
    for i in range(year_index):
        year = 2000 + i
        try:
            year_inflation = data[(data['Year'] == year) & (data['State'] == state)]['Annualized State Inflation Rate'].iloc[0] + 1
        except:
            year_inflation = 1.02056063125 # florida 2014 inflation rate, 2014 florida missing from dataset
        cpi = cpi * year_inflation
    return cpi
    
```

```python
def get_state_real2000(data, column):
    base = 100
    current = data['CPI']
    real_amount = round(data[column] * (base/current), 2)
    return real_amount

def get_state_real2017(data, column, dataset):
    state = data['State']
    try:
        base = dataset[(dataset['Year'] == 2017) & (dataset['State'] == state)]['CPI'].iloc[0]
        current = data['CPI']
    except:
        return np.nan
    real_amount = round(data[column] * (base/current), 2)
    return real_amount
```

```python
master['Real Average WBA (state, 2000 base)'] = master.apply(lambda row: get_state_real2000(row, "Average WBA"), axis=1)
master['Real Maximum WBA (state, 2000 base)'] = master.apply(lambda row: get_state_real2000(row, "Maximum WBA"), axis=1)
master['Real Minimum WBA (state, 2000 base)'] = master.apply(lambda row: get_state_real2000(row, "Minimum WBA"), axis=1)
master['Real Average WBA (state, 2023 base)'] = master.apply(lambda row: get_state_real2017(row, "Average WBA", master), axis=1)
master['Real Maximum WBA (state, 2023 base)'] = master.apply(lambda row: get_state_real2017(row, "Maximum WBA", master), axis=1)
master['Real Minimum WBA (state, 2023 base)'] = master.apply(lambda row: get_state_real2017(row, "Minimum WBA", master), axis=1)
master.head()
```
